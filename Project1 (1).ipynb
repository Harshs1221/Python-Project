#STATE

```python
df.groupby('State')['Orders'].sum()
```

```python
sale_state = df.groupby('State',as_index= False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data = sale_state, x = 'State',y = 'Orders',palette='viridis', hue = 'State')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title("Orders by Top 10 States")
plt.show()
```

from above graph we can see that U.P, Maharashtra, Karnataka are the top 3 states in which most orders are placed

```python
oc = df.groupby("Occupation", as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False)

sns.set(rc={'figure.figsize':(20,7)})
ax = sns.barplot(data = oc, x = 'Occupation', y = 'Orders', palette='magma',hue = "Occupation")
ax.set_title("Orders by Occuptaion",fontsize=20)
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
oa = df.groupby('Occupation',as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending= False)

ax = sns.barplot(data = oa, x = 'Occupation', y = 'Amount', palette='magma',hue = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title("Amount by Occupation", fontsize = 20)

plt.show()
```

from above graph, we can see that most amount are spend by IT Sector, Healthcare and Aviation

PRODUCT CATEGORY

```python
df.head(1)
```

```python
#top products by orders
p = df.groupby("Product_Category", as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending= False)

ax= sns.barplot(data = p, x = "Product_Category", y = 'Orders', palette='crest',hue = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title("Orders by Product Category", fontsize = 20)
plt.xticks(rotation=45, ha='right')
plt.show()
```

from above graph, we can see that most order are for clothing & apparel, Food and Electronics & gadgets.

```python
pa = df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending= False)

ax= sns.barplot(data = pa, x = "Product_Category", y = 'Amount', palette='flare', hue = 'Product_Category')

ax.set_title("Amount By Product category", fontsize= 20)
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

from above graph, we can see that most amount are from Food, Clothing & Apparel and Electronics.
